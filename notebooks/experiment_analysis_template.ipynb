# Sweep Analysis Template

Analyze local sweep summaries, training logs, and evaluation logs without machine-specific paths.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 140)

PROJECT_ROOT = Path.cwd().resolve().parent
RESULTS_DIR = PROJECT_ROOT / "results"
SWEEPS_DIR = RESULTS_DIR / "sweeps"

TRAIN_LOG_PATH = RESULTS_DIR / "train_log_v2.csv"
EVAL_LOG_PATH = RESULTS_DIR / "evaluate_log_v2.csv"

AVAILABLE_SWEEP_CSVS = sorted(SWEEPS_DIR.glob("*.csv"))
AVAILABLE_SWEEP_CSVS
```

```python
if not AVAILABLE_SWEEP_CSVS:
    raise FileNotFoundError("No sweep CSVs found under results/sweeps")

SWEEP_CSV_PATH = AVAILABLE_SWEEP_CSVS[-1]
SWEEP_CSV_PATH
```

```python
summary_df = pd.read_csv(SWEEP_CSV_PATH)
train_df = pd.read_csv(TRAIN_LOG_PATH) if TRAIN_LOG_PATH.exists() else pd.DataFrame()
eval_df = pd.read_csv(EVAL_LOG_PATH) if EVAL_LOG_PATH.exists() else pd.DataFrame()

print(f"sweep rows: {len(summary_df)}")
print(f"train rows: {len(train_df)}")
print(f"eval rows: {len(eval_df)}")
print(f"sweep columns: {list(summary_df.columns)}")

summary_df.head()
```

```python
leaderboard = summary_df.sort_values(["eval_mAP", "recall@1"], ascending=False)
leaderboard[[c for c in ["combo_name", "embedding_dim", "learning_rate", "sampler_p", "sampler_k", "eval_mAP", "recall@1", "best_epoch"] if c in leaderboard.columns]]
```

```python
plot_df = leaderboard.head(10).copy()
plt.figure(figsize=(12, 5))
sns.barplot(data=plot_df, x="combo_name", y="eval_mAP")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
