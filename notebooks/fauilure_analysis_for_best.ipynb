# Failure Analysis for Best Checkpoint

Load the current best checkpoint, inspect failure-case CSVs, and visualize sampled failure examples.

```python
import random as rand
import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent
SRC_DIR = PROJECT_ROOT / "src"

if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from evaluate import Evaluate
from visualise import query

RUN_ID = "20260405_012600"
CHECKPOINT_PATH = PROJECT_ROOT / "checkpoints" / RUN_ID / "best.pth"
RECALL_AT1_FAILCASE = PROJECT_ROOT / "results" / RUN_ID / "best_fail_cases_recall_at_1.csv"
RECALL_AT5_FAILCASE = PROJECT_ROOT / "results" / RUN_ID / "best_fail_cases_recall_at_5.csv"
RECALL_AT10_FAILCASE = PROJECT_ROOT / "results" / RUN_ID / "best_fail_cases_recall_at_10.csv"

eval_obj = Evaluate(CHECKPOINT_PATH)
```

```python
fail_case1 = pd.read_csv(RECALL_AT1_FAILCASE)
fail_case5 = pd.read_csv(RECALL_AT5_FAILCASE)
fail_case10 = pd.read_csv(RECALL_AT10_FAILCASE)
```

```python
fail_case = fail_case1
fail_case.head()
```

```python
paths = fail_case["fail_case"].tolist()
PICK_CHECK_NUM = min(20, len(paths))
check_list = rand.sample(paths, PICK_CHECK_NUM)

for case_path in check_list:
    query(case_path, CHECKPOINT_PATH, 5, eval=eval_obj)
```
